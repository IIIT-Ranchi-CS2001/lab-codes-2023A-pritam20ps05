```python
import pandas as pd
import numpy as np

# Loading data into pandas dataframe
df = pd.read_csv('AQI_Data.csv')

# Q(a)
# Display first 5 rows
df.head(5)
```

```python
# Q(b)
# Display last 6 rows
df.tail(6)
```

```python
# Q(c)
# Display summary statistics for all numeric columns
df.describe()
```

```python
# Q(d)
# Computing mean of AQI PM2.5 PM10 values for each city
mean_values_per_city = df.groupby("City")[["AQI", "PM2.5", "PM10"]].apply(
    lambda group: np.mean(group, axis=0)
)

print(mean_values_per_city)
```

```python
# Set - 2
# Q(2a)
# Check and fill missing data in numeric columns with column mean
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
for column in numeric_columns:
    if df[column].isnull().any():
        df[column].fillna(df[column].mean(), inplace=True)

print(df)
```

```python

# Q(2b)
# Display mean median and Standard Deviation
aqnp = df.get('AQI').to_numpy()
print(aqnp.mean())
print(np.median(aqnp))
print(aqnp.std())
```
